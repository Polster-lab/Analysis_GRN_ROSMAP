# 6. Logistic Regression

```python
import pandas as pd
import random
from sklearn.decomposition import PCA
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
from sklearn.preprocessing import StandardScaler


# Assuming your DataFrame contains only numeric values
# If not, you might need to preprocess your data accordingly



# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(outlier_corrected_combined.drop(columns='Label'), outlier_corrected_combined['Label'], test_size=0.2, random_state=42)
#y_train = random.sample(y_train.tolist(), len(y_train.tolist()))
#y_test = random.sample(y_test.tolist(), len(y_test.tolist()))
# Initialize and train the logistic regression model
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
model = LogisticRegression(class_weight={'AD':1,'NCI':1.42})
model.fit(X_train_scaled, y_train)


# Make predictions on the test set
y_pred = model.predict(X_test_scaled)

# Evaluate the model performance
accuracy = accuracy_score(y_test, y_pred)
conf_matrix = confusion_matrix(y_test, y_pred)
class_report = classification_report(y_test, y_pred)

print(f"Accuracy: {accuracy:.2f}")
print("\nConfusion Matrix:")
print(conf_matrix)
print("\nClassification Report:")
print(class_report)
```

# 7. Bootstrap by randomly assigning labels

```python
import pandas as pd
import random
from sklearn.decomposition import PCA
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
from sklearn.preprocessing import StandardScaler


# Assuming your DataFrame contains only numeric values
# If not, you might need to preprocess your data accordingly

accuracies = []

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(outlier_corrected_combined.drop(columns='Label'), outlier_corrected_combined['Label'], test_size=0.2, random_state=42)

for i in tqdm(range(1000)):

    y_train = random.sample(list(y_train), len(list(y_train)))
    y_test = random.sample(list(y_test), len(list(y_test)))
    # Initialize and train the logistic regression model
    scaler = StandardScaler()
    X_train_scaled = scaler.fit_transform(X_train)
    X_test_scaled = scaler.transform(X_test)
    model = LogisticRegression(class_weight={'AD':1,'NCI':1.42})
    model.fit(X_train_scaled, y_train)

    # Make predictions on the test set
    y_pred = model.predict(X_test_scaled)

    # Evaluate the model performance
    accuracy = accuracy_score(y_test, y_pred)
    conf_matrix = confusion_matrix(y_test, y_pred)
    class_report = classification_report(y_test, y_pred)
    accuracies.append(accuracy)

print(f"Accuracy: {accuracy:.2f}")
print("\nConfusion Matrix:")
print(conf_matrix)
print("\nClassification Report:")
print(class_report)
```

```python
import numpy as np
confidence_interval = np.percentile(accuracies, [2.5, 97.5])
confidence_interval
```
