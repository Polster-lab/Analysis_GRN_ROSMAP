```python
import scipy.stats
```

```python
x = []
y = []
p = []
from tqdm import tqdm
for index, row in tqdm(NCI_Plaques_Gene_Gene.iterrows()):
    NCI_list = list(row.values)
    AD_list = list(AD_Plaques_Gene_Gene.loc[index].values)
    
    import scipy.stats as stats



# Perform Mann-Whitney U test
    statistic, p_value = stats.mannwhitneyu(NCI_list, AD_list)

    # Set a significance level (alpha)
    alpha = 0.05

    # Check if the p-value is less than alpha
    if p_value < alpha:
        x.append(1)
        y.append(index)
        p.append(p_value)
    else:
        x.append(0)
```

```python
statistically_significant_gene_gene = pd.DataFrame([y,p]).T
statistically_significant_gene_gene.set_index(0, inplace = True)
statistically_significant_gene_gene = statistically_significant_gene_gene.sort_values(by = 1).rename(columns={1:'P value'})
statistically_significant_gene_gene.head()
```

# Go Enrichment Analysis

## 1. Gene_gene

```python
from genes_ncbi_9606_proteincoding import GENEID2NT as GeneID2nt_human
from goatools.base import download_go_basic_obo
from goatools.base import download_ncbi_associations
from goatools.obo_parser import GODag
from goatools.anno.genetogo_reader import Gene2GoReader
from goatools.goea.go_enrichment_ns import GOEnrichmentStudyNS
```

```python
top_statistically_significant_gene_gene = list(statistically_significant_gene_gene['0'])
```

```python
l = []
for i in top_statistically_significant_gene_gene:
    
    l.extend(i.split('_'))
l = list(set(l))
```

```python
obo_fname = download_go_basic_obo()
fin_gene2go = download_ncbi_associations()
obodag = GODag("go-basic.obo")
```

```python
mapper = {}

for key in GeneID2nt_human:
    mapper[GeneID2nt_human[key].Symbol] = GeneID2nt_human[key].GeneID
    
inv_map = {v: k for k, v in mapper.items()}
```

```python
objanno = Gene2GoReader(fin_gene2go, taxids=[9606])
# Get namespace2association where:
#    namespace is:
#        BP: biological_process               
#        MF: molecular_function
#        CC: cellular_component
#    assocation is a dict:
#        key: NCBI GeneID
#        value: A set of GO IDs associated with that gene
ns2assoc = objanno.get_ns2assc()
```

```python
#run one time to initialize
goeaobj = GOEnrichmentStudyNS(
        GeneID2nt_human.keys(), # List of mouse protein-coding genes
        ns2assoc, # geneid/GO associations
        obodag, # Ontologies
        propagate_counts = False,
        alpha = 0.05, # default significance cut-off
        methods = ['fdr_bh'])
```

```python
#run one time to initialize
GO_items = []

temp = goeaobj.ns2objgoea['BP'].assoc
for item in temp:
    GO_items += temp[item]
    

temp = goeaobj.ns2objgoea['CC'].assoc
for item in temp:
    GO_items += temp[item]
    

temp = goeaobj.ns2objgoea['MF'].assoc
for item in temp:
    GO_items += temp[item]
```

```python
#pass list of gene symbols
def go_it(test_genes):
    print(f'input genes: {len(test_genes)}')
    
    mapped_genes = []
    for gene in test_genes:
        try:
            mapped_genes.append(mapper[gene])
        except:
            pass
    print(f'mapped genes: {len(mapped_genes)}')
    
    goea_results_all = goeaobj.run_study(mapped_genes)
    goea_results_sig = [r for r in goea_results_all if r.p_fdr_bh < 3]
    GO = pd.DataFrame(list(map(lambda x: [x.GO, x.goterm.name, x.goterm.namespace, x.p_uncorrected, x.p_fdr_bh,\
                   x.ratio_in_study[0], x.ratio_in_study[1], GO_items.count(x.GO), list(map(lambda y: inv_map[y], x.study_items)),\
                   ], goea_results_sig)), columns = ['GO', 'term', 'class', 'p', 'p_corr', 'n_genes',\
                                                    'n_study', 'n_go', 'study_genes'])

    #GO = GO[GO.n_genes > 1]
    return GO
```

```python
import mygene
mg = mygene.MyGeneInfo()
geneList = set(l)
geneSyms = mg.querymany(geneList , scopes='ensembl.gene', fields='symbol', species='human')
k = []
for i in geneSyms:
    
    try:
        k.append(i['symbol'])
    except:
        continue
```

```python
df = go_it(k)
```

```python
df['class'].value_counts()
```

```python
biological_process_df = df[df['class']== 'biological_process']
biological_process_df = biological_process_df[biological_process_df.p_corr <0.05]
```

```python
biological_process_df.term.tolist()
```

```python
molecular_function_df = df[df['class']== 'molecular_function']
molecular_function_df = molecular_function_df[molecular_function_df.p_corr <0.05]
```

```python
molecular_function_df.term.tolist()
```

```python
cellular_component_df = df[df['class']== 'cellular_component']
cellular_component_df = cellular_component_df[cellular_component_df.p_corr <0.05]
```

```python
cellular_component_df.term.tolist()
```

```python
import plotly.express as px
molecular_function_df.sort_values(by = 'p_corr', ascending= False,inplace= True)
fig = px.bar(molecular_function_df, x='p_corr', y = 'term', orientation = 'h')
fig.update_layout(xaxis = dict(title = 'Corrected P value for multiple testing'))
fig.update_layout(width = 1500, height = 500)
fig.update_layout(title= dict(text = 'Go Enrichment Analysis (Molecular Function)', x = 0.5))
fig.show()
```
