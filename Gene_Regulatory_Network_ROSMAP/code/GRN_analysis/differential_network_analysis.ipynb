```python


from netZooPy import condor
```

```python
differential_df = merge_df[['tf','gene','force_change_diff']]
```

```python
list_of_tf = list(set(differential_df.tf))
differential_df = differential_df[~differential_df.gene.isin(list_of_tf)]
```

```python
differential_df.rename(columns={'tf':'pollinator','gene':'plant','force_change_diff':'interactions'})
```

```python
condor_object = condor.condor_object(dataframe=differential_df)
```

```python
condor_object.initial_community()
```

```python
condor_object.brim(deltaQmin="def")
```

```python
condor_object.reg_memb.head(5)
```

```python
condor_object.tar_memb.head(5)
```

```python
def fun(s):
    return s.split('_')[1]

x = condor_object.tar_memb.tar.apply(fun)
condor_object.tar_memb.tar = x
```

```python
condor_object.tar_memb.head()
```

```python
x = condor_object.reg_memb.reg.apply(fun)
condor_object.reg_memb.reg = x
condor_object.reg_memb.head()
```

```python
m = []
for i in differential_df.tf:
    m.append(condor_object.reg_memb[condor_object.reg_memb.reg==i]['community'])
    
```

```python
differential_df['tf_cluster_no'] = m
```

```python
m = []
for i in differential_df.gene:
    m.append(condor_object.tar_memb[condor_object.tar_memb.tar==i]['community'])
```

```python
differential_df['gene_cluster_no'] = m
```

```python
differential_df['tf_cluster_no'] = differential_df['tf_cluster_no'].astype(int)
```

```python
differential_df['gene_cluster_no'] = differential_df['gene_cluster_no'].astype(int)
```
